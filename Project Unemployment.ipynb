```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your data is in a CSV file named 'unemployment_data.csv'
df = pd.read_csv(r'D:\Project\Unemployment in India.csv')

# **Correct the column name (assuming it's 'Estimated Labour Participation Rate')**
unemployment_rates = df['Area'].value_counts()
unemployment_rates.plot(kind='pie', autopct='%1.1f%%')  # Format percentage with one decimal place

plt.title('Distribution of Areas', weight='bold', fontsize=14)
plt.legend(loc='upper right', fontsize=12, shadow=True, bbox_to_anchor=(2, 1))  # Set legend properties
plt.show()
```

```python
#Bar Chart - Unemployment Rate by Region
sns.barplot(x='Region', y='Estimated Unemployment Rate', data=df)  # Replace 'Region' if needed
plt.title('Unemployment Rate by Region', weight='bold', fontsize=14)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability with many regions
plt.xlabel('Region')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```

```python
#Scatter Plot - Unemployment Rate vs. Labor Participation Rate 
sns.scatterplot(x='Estimated Unemployment Rate', y='Estimated Labour Participation Rate', data=df)
plt.title('Unemployment Rate vs. Labor Participation Rate', weight='bold', fontsize=14)
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Labor Participation Rate (%)')
plt.show()
```

```python
# Histogram - Frequency Distribution of Unemployment Rates
sns.histplot(df['Estimated Unemployment Rate'])  # Replace column if needed
plt.title('Distribution of Unemployment Rates', weight='bold', fontsize=14)
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
#Heatmap - Unemployment Rate by Region and Date 
pivot_table = df.pivot_table(values='Estimated Unemployment Rate', index='Region', columns='Date')  # Replace columns if needed
sns.heatmap(pivot_table, cmap='coolwarm')  # Adjust 'cmap' for different color schemes
plt.title('Unemployment Rate by Region and Date', weight='bold', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Region')
plt.show()
```

```python
#Bar plot visualization code for state wise Estimated Employed Workforce
# Sort data by 'Estimated Employed' in descending order
df_sorted = df.sort_values(by='Estimated Employed', ascending=False)

# Create a bar plot
sns.barplot(x='States', y='Estimated Employed', data=df_sorted)

# Customize the plot
plt.title('State-wise Estimated Employed Workforce', weight='bold', fontsize=12)
plt.xlabel('States', fontsize=10)
plt.ylabel('Estimated Employed', fontsize=10)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust spacing for better visualization
plt.show()
```
